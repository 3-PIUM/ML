```python
import pandas as pd
import os

file_list = [
    "D:/PIUM/데이터_전처리/data/클렌징/상품정보/클렌징_립&아이리무버_상품정보.csv",
    "D:/PIUM/데이터_전처리/data/클렌징/상품정보/클렌징_오일-밤_상품정보.csv",
    "D:/PIUM/데이터_전처리/data/클렌징/상품정보/클렌징_워터-밀크_상품정보.csv",
    "D:/PIUM/데이터_전처리/data/클렌징/상품정보/클렌징_클렌징디바이스_상품정보.csv",
    "D:/PIUM/데이터_전처리/data/클렌징/상품정보/클렌징_클렌징폼-젤_상품정보.csv",
    "D:/PIUM/데이터_전처리/data/클렌징/상품정보/클렌징_티슈-패드_상품정보.csv",
    "D:/PIUM/데이터_전처리/data/클렌징/상품정보/클렌징_필링&스크럽_상품정보.csv"
]

df_list = []

for file_path in file_list:
    # 파일명에서 카테고리만 추출
    filename = os.path.basename(file_path)
    category = filename.split("_")[1]  # '립메이크업' / '베이스메이크업' 등

    print(f"불러오는 파일: {filename} → category: {category}")

    # CSV 읽기
    df = pd.read_csv(file_path, encoding='utf-8', low_memory=False)

    df = df.drop(columns=["옵션개수","옵션리스트"])

    # 카테고리 열 추가
    df["category"] = category

  

    # 리스트에 추가
    df_list.append(df)

# 병합
merged_df = pd.concat(df_list, ignore_index=True)

# 열 순서 재정렬: '카테고리'를 맨 앞으로
cols = ["category"] + [col for col in merged_df.columns if col != "category"]
merged_df = merged_df[cols]


merged_df
```

## 상품별 바코드 생성

```python
import re
import os
from barcode import EAN13
from barcode.writer import ImageWriter

# 바코드 저장 폴더
barcode_image_folder = r"D:/PIUM/데이터_전처리/data/클렌징/바코드"
os.makedirs(barcode_image_folder, exist_ok=True)

# 카테고리 코드 매핑
# 메이크업
category_dict = {
    "클렌징폼-젤": "100000100100001",
    "오일-밤": "100000100100004",
    "워터-밀크": "100000100100005",
    "필링&스크럽": "100000100100007",
    "티슈-패드":"100000100100008",
    "립&아이리무버":"100000100100006",
    "클렌징 디바이스":"100000100100009"
}

COUNTRY_CODE = "880"

# 상품코드 추출 함수
def extract_goods_code(image_url):
    match = re.search(r"[AB]\d{10,15}", str(image_url))
    if match:
        full_code = match.group(0)[1:]  # A 또는 B 제거
        if len(full_code) >= 8:
            return full_code[-8:-2]
    return None

# EAN13 체크섬 계산 함수
def calculate_ean13_checksum(number12):
    digits = list(map(int, number12))
    odd_sum = sum(digits[::2])
    even_sum = sum(digits[1::2])
    checksum = (10 - ((odd_sum + even_sum * 3) % 10)) % 10
    return str(checksum)

# 바코드, 상품분류코드 생성 및 DataFrame에 삽입
def add_barcodes_to_all_masks(df):
    barcodes = []

    for idx, row in df.iterrows():
        category = row.get("category", "").strip()
        category_code = category_dict.get(category)

        if not category_code:
            print(f"[{idx}] ⚠️ 알 수 없는 카테고리: {category}")
            barcodes.append("INVALID")
            continue

        goods_code = extract_goods_code(row.get("이미지", ""))
        if not goods_code:
            barcodes.append("INVALID")
            continue

        major_code = category_code[9:11]
        minor_code = category_code[-1:]
        product_code_suffix = goods_code[-6:].zfill(6)

        ean12 = COUNTRY_CODE + major_code + minor_code + product_code_suffix
        ean12 = ean12[:12]
        checksum = calculate_ean13_checksum(ean12)
        full_ean13 = ean12 + checksum

        try:
            barcode_obj = EAN13(ean12, writer=ImageWriter())
            barcode_path = os.path.join(barcode_image_folder, f"{full_ean13}")
            barcode_obj.save(barcode_path)
            barcodes.append(full_ean13)
        except Exception as e:
            print(f"[{idx}] 바코드 생성 오류: {e}")
            barcodes.append("ERROR")

    df["바코드"] = barcodes
    return df

# 실제 실행
merged_df = add_barcodes_to_all_masks(merged_df)

# INVALID 바코드 개수
invalid_count = (merged_df["바코드"] == "INVALID").sum()
total_count = len(merged_df)
invalid_ratio = invalid_count / total_count * 100

print(f"INVALID 바코드 수: {invalid_count}개")
print(f"전체 대비 비율: {invalid_ratio:.2f}%")
```

```python
# "INVALID" 또는 "ERROR" 바코드 제거
merged_df = merged_df[~merged_df["바코드"].isin(["INVALID", "ERROR"])].reset_index(drop=True)
merged_df
```

### 카테고리별 상품분류 코드 생성

```python
def generate_product_category_code(category):
    code = category_dict.get(category)
    if code:
        return code[9:11] + code[-2:]
    return "INVALID"

# 열 생성
merged_df["상품분류코드"] = merged_df["category"].apply(generate_product_category_code)
merged_df
```

### 가격 결측값 처리

```python
# 정가가 없고 할인가만 있는 경우 -> 할인가에 있는 값이 정가임
merged_df['정가'] = merged_df.apply(
    lambda row: row['할인가'] if pd.isna(row['정가']) and pd.notna(row['할인가']) else row['정가'],
    axis=1
)
merged_df
```

### 성분 nan 처리

```python
l = len(merged_df[merged_df['성분'].isna()])
print(f"'성분' 열의 NaN 값 개수: {l}개 행 삭제")
merged_df = merged_df[merged_df['성분'].notna()]
merged_df
```

```python
# 쉼표 제거 → 문자열로 처리 → 정수형으로 변환
merged_df["할인가"] = merged_df["할인가"].str.replace(",", "").astype(int)
merged_df["정가"] = merged_df["정가"].str.replace(",", "").astype(int)
merged_df
```

```python
merged_df["할인가"] = merged_df["할인가"].astype(int)
merged_df["정가"] = merged_df["정가"].astype(int)
merged_df
```

```python
merged_df["할인율(%)"] = (1 - merged_df["할인가"] / merged_df["정가"]) * 100
merged_df["할인율(%)"] = merged_df["할인율(%)"].round(1).astype(int)
merged_df
```

### 비건

```python
# '비건'이라는 단어가 포함되면 '비건', 아니면 NaN
merged_df["비건품목"] = merged_df["상품명"].apply(lambda x: 1 if pd.notna(x) and "비건" in x else 0)
merged_df
```
